## 기본 분류

- 정의
  - 데이터를 표현하는(데이터를 담는 그릇)형식
  - 수치(온도,집값,주가 등등...)
  - 문자열(리뷰,댓글,평가 등등...)
  - 데이터의 형태
- 종류
  - 단일 데이터형(1개) >> 데이터 자체(본질)
    - **수치형**
      - 정수 : -1,0,1
      - 부동소수 : 3.14
      - 복소수
    - **문자열**
      - 'Hello'
      - "안녕"
    - **블린형(이진,  0 or 1)**
      - True / False
    - 바이트형
      - 바이너리 데이터
  - 컬렉션(1개 이상)
    -데이터가 1개 혹은 그 이상 모여서 구성된 타입(n개의 단일데이터가 모여서 구성)
    - 종류
      - 파이썬
        - 문자열
          - 단일데이터면서 연속데이터임
        - list(리스트)
          - 순서 존재 >> 인덱스(순서표기)존재 >>정/역방향 인덱스 >>데이터는 인덱스로 접근,관리
        - dict(딕셔너리)
          -순서 없음 >> 키(고유값)와 값(데이터)으로 데이터 관리 >> 인덱스 X >> 키를 통해 값 접근 가능
        - tuple(튜플)
          - 리스트와 거의 동일, 값을 묶는다 읽기 전용(수정 X)
        - set(집합)
          - 중복 데이터 제거
      - 넘파이(수학 과학용 라이브러리)
        - ndarray(배열)
        - **배열내 모든 구성원은 동일타입임**
        - 고속 연산, 수학, 머신/딥러닝등 베이스 기술
      - 판다스(데이터사이언스 용도)
        - Series(1차원)
        - DataFrame(2차원)
        - 데이터 전처리, eda, 분석 등등
      - 딥러닝(파이토치, 텐서플로우 >> 딥러닝엔진(프레임웍) )
        - Tensor(텐서)
        - 텐서단위로 데이터 관리?
        - 모델 학습,활용
      - 머신러닝, LLM, 생성형모델 등은 위의 자료구조 내에서 모두 학습/활용

#  흐름제어
