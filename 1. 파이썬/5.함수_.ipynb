### map()

- 목적
  - 연속된 데이터( 컬렉션 계열들중 리스트 등...)의 멤버들을 하나씩 꺼내서 뭔가 작업( 2배로 확대, 전처리등등)
  - 데이터를 하나씩 꺼내 조작(전처리)
  


````python
datas = [1,2,3,4,5]
datas
````

````python
# 요구사항 datas의 모든 멤버(데이터)들의 값을 2배로 업그레이드 하시오
# [1,2,3,4,5] >> [2,4,6,8,10]
datas * 2 # 의도와 다름
````

````python
# double 함수  << 이런 역할을 담당하는 함수( 뉘앙스 >> 타 언어에서는 콜백함수)
def double (x): # x에는 1,2,3,4,5가 차례대로 세팅되어서 호출됨
  print(f'x={x} >>>> {x*2}')
  return x*2
# 값이 2배로 확대되어서 새로운 리스트 구성
# 비슷한 역할 담당 >> pandas(데이터 사이언스용 라이브러리)의 apply()
list( map(double, datas) ), datas
````

````python
%%time
# datas의 원본 데이터를 3배로 확장하여 다시 리스트 구하시요
def triple (x): # x에는 1,2,3,4,5가 차례대로 세팅되어서 호출됨
  print(f'x={x} >>>> {x*3}')
  return x*3

list( map(triple, datas) ), datas
````

````python
# 리스트 컴프리핸션을 통해서 동일한 결과를 구하시오 (실습) >> 코드를 줄여라/압축해라
datas = [1, 2, 3, 4, 5]
triple_datas_comprehension = [x * 3 for x in datas]
triple_datas_comprehension
````

````python
%%time
# 결과중심 구성
[v*3 for v in datas]
````

### filter()

````python
# 빈도 계산
datas = "sdjahskjshfkjsdhfjknasjknvjkvjksanksfjksndfjksdnvjksankvlsdvkjnasjkndsjnvsjkdanvjksnvjknsvjfvkfsjkvnjfsknvjsksndnjdnfj"
datas

#요구사항
# datas가 참조하는 물자열을 대상으로 알파벳 문자의 개별 빈도를 카운트하여 26개의 방을 가진 리스트에 담으시오
# 단 a는 0번방, b는 1번, z는 25번으로 순서대로 배치함
# 라이브러리 사용 배제
# 1. 빈도를 담을 그릇 필요 >> 공간은 26개 (알파벳 문자의 총수)
counts = [0] *26
# 2. 말뭉치(코퍼스)로부터 문자 1개씩 추출 >> 마지막 문자까지 반복 >> for문
for ch in datas: # 문자열을 for문으로 처리 >> 문자 1개씩 차례대로 추출됨
  # 2-1. 문자 >> ord() 처리 >> 문자의 axcii코드 추출 >> 방번호 체킹 (해당 문자의 ascii - a 의 ascii) >> +1
  #print(ch, ord(ch), ord(ch) - ord('a'))
  # 특정 문자 빈도 증가
  counts[ord(ch) - ord('a')] += 1
  #break

# 빈도를 담은 리스트값 출력, 원데이터(말뭉치)문자수 == 각 문자의 빈도 총합
print( counts ), len(datas), sum(counts)
````

````python
# 정규화 >> 모든 빈도를 0과 1사이로 재구성
# 개별빈도/전체빈도 >> 비중(특정국가의 특정문자 사용 비중)
total_freq = sum(counts) # 전체빈도
list(map(lambda x:x/total_freq, counts) )
# 차후 머신러닝/딥러닝 에서는 softmax() : 활성화함수로 간단하게 처리됨 >> 비중체크
````

# scope : 변수의 범위

- 기본
  - 함수 내부에서 정의된 변수는 **지역(로컬)변수**라고 함
  - 절차적 코드에서 맨 앞칸에 정의된 변수 >> 전역(글로벌)변수
- 종류
  - 전역변수 : global
    - 코드 전방위로 사용 가능
  - 지역변수 : local
    - 함수 내에서만 의미를 가짐
    - 함수 밖에서는 존재/사용 X
  - 넌지역변수 : nonlocal
    - 함수안내 함수가 존재하는 구조속에서 사용됨


````python
# 전역변수
score = 100
# 함수
def test() :
  # 함수 내에서 전역변수 사용
  print (score)

# 사용
test()
print( '->', score)
````

````python
# 전역변수
score = 100
# 함수
def test() :
  # 함수내의 score는 지역(O), 전역(X) << 이름이 전역변수와 동일한 지역변
  score = 0
  score +=1
  print (score)

# 사용
test()
print( '->', score)
````

````python
# 전역변수
score = 100
# 함수
def test() :
  # 굳이 전역변수를 함수 내부에서 엑세스(읽고,쓰기(수정))
  # score는 전역변수야! 라는 명시적표현 필요
  global score
  score +=1
  # 전역변수를 함수 내부에서 조작했다.
  print (score)

# 사용
test()
print( '->', score)
````
