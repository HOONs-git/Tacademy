````python
import random
#stage 4
ai_number = random.randint(1,100)
#시도 횟수 >> 1~100 사이 값이 들어가서 판단이 진행될때 카운트됨
try_count = 0

# stage 6
while True:
  # stage 5
  while True:
    player_number=input('AI가 생성한 값 1~100 사이중 하나의 정수값을 맞추시오 \n').strip()

    if not player_number: # 값이 비어있다
      print('값이 비어있습니다. 정확하게 입력하세요')
      continue

    if not player_number.isnumeric():
      print('정수값만 입력가능합니다. 정확하게 입력하세요.')
      continue

    player_number = int(player_number)

    if (player_number < 1) or (player_number > 100):
      print("정확하게 입력하세요")
      continue
    break

    # stage6  실제 구현 내용 >> 판단
  try_count += 1
  if player_number > ai_number:
    print('힌트 - 입력한게 더 크다')
    pass
  elif player_number < ai_number:
    print('힌트 - 입력한게 더 작다')
    pass
  else:
    print('정답')
    try_count -= 1 # 설정: 시도횟수를 차감(보정) >> 1번에 맞췄을떄 100점가능
    break # stage 6 탈출
# 추가 설정 기획(요구사항상) 10회 이상 진행해도 못맞추면에 대한 컨셉 없음
print( f'{ (10 - try_count) * 10}점 획득 하였습니다.')
````
