# 조건식

- 특징
  - 조건문, 반복문이 구동(실행)되게 하는 기본 전제 조건
- 결론
  - 조건식의 결론 >> True, False (조건|반복문 내부 진입 X) >> 블린형
  - False로 인식하는 표현
    - None, '', 0, 0.0, [], {}, (), False
- 연산자
  - 값 vs 값 비교하는 조건식
    ```
    A == B : A와B는 값이 일치하는가?
    A != B : A와B는 값이 다른가?
    A > B : A는 B보다 큰가? >> 초과
    A < B : A는 B보다 작은가? >> 미만
    A >= b : A는 B보다 크거나 같은가? >> 이상
    A <= B : A는 B보다 작거나 같은가? >> 이하
    not A : A의 갑을 부정한다
      - 부정상황을 체킹할떄 사용
      - 부정상황을 부정(not) >> True가 감지되면 >> A값은 비어있다(False상황이다)

    ```
  - 조건식 vs 조건식을 비교하는 연산다
    - 타 언어에서는 쇼트연산자라고 부름
    ```
      # (조건식=C1) 연산자 (조건식=C2)
      # 조건이 복잡한 경우 빠른 처리를 위함, 불필요한 연산 제거

      **C1 or C2**
         - **C1이 참이면 그대로 조건 체크 종료**
         - C2는 체크할 필요 없음, C2의 결과에 상관없이 무조건 참
         - C1이 거짓이면 C2까지 체크하고 결론 내림
         - ~ 이거나 ~
      **C1 and C2**
        - **C1이 거짓이면 그대로 조건체크 종료**
        - C2는 체크할 필요 없음. C2의 결과에 상관없이 무조건 참
        - C1이 참이면 C2까지 체크하고 결론 내림
          - 참의 결론에 도달하기 위해서는 반드시 C1, C2가 모두 참
          - ~ 이고 ~
    ```
  - 기타
    ```
      A is B
        - A는 B인가? >> 객체를 관리하는 내부 id를 비교하는 연산자
        - 동일 객체인가?(본질이 같은가? 메모리상에서)
      a in A
        - A안에 멤버로 a가 존재하는가?
    ```

- syntax (문법의 구조)
  ```
  #상황(조건, 컨디션) 1개

    수행문(statements) or pass( 수행X, 자리는 차지 )
  #상황(조건, 컨디션) 2개
    if 조건식:
      수행문(statements)
    else:
      수행문(statements)
  #상황(조건, 컨디션) 2개 이상
    if 조건식:

      수행문(statements)
    ...
    else:
      수행문(statements)
  ```

````python
from os import copy_file_range
# 요구사항
  # 이상한 pos기 기계의 결재하고 가정
  # 결제에 따른 메세지 전송
# 조건
  # 체크카드(잔액 : 5천원)의 잔액에 따라 아래 형태가 작동된다고 가정한다.
  # 총 결제액이 5천원 초과 (>)
    # 메세지 : X원이 부족합니다
  # 총 결제액이 5천원 미만 (<)
    # 메세지 : X원이 남았습니다
  # 총 결제액이 5천원 일치 (==)
    # 메세지 : 딱 맞게 결제 하였습니다.


# 고객 체크카드의 잔액
customer_money = 5000
# 임시 설정
while True:
  # 주문 총액을 입력받음 >> 문자열 >> 공백제거 >> 정수변환 >> 변수세팅
  # input() : 내장함수, 사용자 입력 무한대기(블럭킹 코드진행 X)
  # 터미널, 표준입력, 콘솔등에서 입력대기
  # 코랩 등 에서는 웹환경이므로, 입력창으로 대체됨
  # 사용자가 입력후 엔터키를 입력하면 종료
  coffe_price = int( input('구매 총액을 입력하세요 : \n').strip() ) # strip(): 공백제거
  print( coffe_price, type(coffe_price) )

  if coffe_price > customer_money: # 조건식1:
    #print(coffe_price - customer_money,'원이 부족합니다' )
    # 문자열 포멧팅을 통해서 동적으로 메세지 구성
    print( f'{coffe_price - customer_money}원이 남았습니다')
  elif coffe_price < customer_money: # 조건식2:
    #print(customer_money - coffe_price,'원이 남았습니다')
    #customer_money = customer_money - coffe_price
    print( f'잔액 {customer_money - coffe_price} 원이 남았습니다')
  else: # coffe_price == customer_money
    print('딱 맞게 결제하였습니다.')
    break

  # 반복분 탈출코드
  # coffe_price의 값이 0과 같으면 break() 수행
  if coffe_price == 0:
    break # 브레이크를 감싸고 있는 모든 반복문들중 가장 가까운 반복문을 탈출한다.

# 실습 >> 구매 총액에 따른 메세지 출력 하시오
````

### 리스트 + 딕셔너리 형태의 반복문

````python

ktx_stations = [
  {
    'name' : '영등포역',
    'age' : 70,
    'flag' : [1,2,3]
  },
  {
      'name' : '광명역',
      'age' : 25,
      'flag' : [5,6,7]
  }
]
ktx_stations
````

````python
# 요구사항
'''
  영등포역의 년식은 70년
  광명역의 년식은 25년
'''

for ktx_station in ktx_stations:
  # ['영등포역', 70, [1, 2, 3] ] >> 변수에 직접 할당
  name, age, _ = ktx_station.values()
  print( f'{name}의 년식은 {age}년' )
````

### break, continue, else


- 반복문을 중단(탈출)
  - **break를 감싸고 있는 가장 가까운 반복문 탈출**
- continue
  - continue를 감싸고 있는 가장 가까운 반복문으로 점프 >> 다음번 반복문 진행
- else
  - for ~ else ~
  - while ~ else ~
  - 반복문이 중단없이 정상적으로 모두 반복행위를 하고 마무리 되었다면 진입 >> 강제로 반복문이 종료되었는지 점검할수 있다.
  - **필요하면 사용**
  - 어떤 변수없이 그대로 반복문(혹은 예외처리)진행되면 진입>> 뉘앙스

#### break

````python
# 목적
# 임시적 코드, 샘플링을 위해 사용, 1회성 작동
# 한번만 돌려보고 판당하기 위한 임시코드
for n in range(1,10):
  print(n)
  break
````

````python
for n in range(10):
  # n이 3을 초과하면 중단

  if n > 3:
    break #break는 통상 조건문과 같이 사용된다 >> 탈출코드
  print(n)
````

````python
for n in range(10):
  print(n)

  if n > 3:
    break # 중단 위치는 요구사항에 맞게 잘 구성함

````

````python
for n in range( 3, 7): #1번라인
  for m in range( 1, 3): #2번라인
    print(f' {n} x {m} = {n*m:>2} ' )
    break # 2번 라인의 for문과 반응
  break # 1라인의 for문과 반응하여 중간
````

#### continue
  - 반복문을 모수 수행은 한다

````python
for n in range(2,6):
  # 짝수값을 제외하고 출력
  if(not n%2): #짝수라면
    continue # 조건을 만족하면 아래코드 수행x 반복문(1번라인 range())으로 점프
  print(n)
````

#### else

````python
for i in range(3):
  print('A', i)
  if i == 1:
    print('B', i)
    continue
  print('C', i)
else:
  print('continue 사용시 else에 진입한다')
````
