### 인덱싱

- 문자열에서 특정 문자열을 **추출** 하는 행위
- 대부분 컬렉션 계열 자료구조에서 지원

  - **차원축소**
    - 2차원 >> 인덱싱 >> 1차원
    - 리스트, 딕셔너리, 투플, 배열, 시리즈, 데이터프레임, 텐서

### 슬라이싱(자른다,범위)

- 원본 데이터에서 특정 데이터만 자른다
- **차원 유지**
  - 문자열을 제외한 다른 자료구조에서 해당
  - 2D -> 슬라이싱 -> 20
- 데이터는 연속적르오 이웃해야한다.(한 반향성)
  - step옵션을 통해서 간격을 조절할수 있따
- 문법
  ```
    변수명[ 시작 인덱스 : 끝 인덱스 : step(간격,기본값1,생략가능함) ]
    or
    변수명[ 시작 인덱스 : 끝 인덱스  ] << 이렇게써도 가능함 간격은 무조건1
    or
    변수명[ : 끝 인덱스  ]  << 이렇게쓰면 처음부터~끝 인덱스까지
    or
    변수명[ 시작 인덱스 :  ]  << 이렇게쓰면 시작인덱스~끝 인덱스까지
    or
    변수명[ :  ]  << 이렇게쓰면 처음부터 끝까지 = copy(복사)
  ```

````python
#카피
a[ : ]
````

````python
#처음부터 2개만 자른다. 앞쪽(prefix)에서 특정 문자열 자를때 유용
# 0 <= 인덱스 < 2
a[ : 2]
````

````python
#맨뒤에서 2개만 자른다
# -2 <= 인덱스 < 끝까지
a[ -2: ]
````

````python
# 앞뒤로 한문자씩 제외한것과 동일
a[ 1: -1], a[ 1: -1: 1]
````

  - 데이터베이스의 sql >> join 구문과 유사함
  - pandas의 merge와 유사성 가짐

````python
# 데이터 준비
a1 = [ 1, 3, 4, 6, 2, 7, 8, 8, 3, 2, 100 ]
a2 = [ 5, 3, 6, 9, 256]

# 중복제거 수행
a1_set = set(a1)
a2_set = set(a2)

a1_set, a2_set
````

````python
# 교집합 >> 2개 집합에서 일치되는 데이터만 추출
a1_set. intersection( a2_set )
````

````python
# 합집합 >> 2개의 집합을 모두 합쳐서, 중복제거
a1_set.union(a2_set)
````

````python
# 차집합 >> a1_set에만 존재하는 데이터만 추출( 고유한 데이터만 추출 )
a1_set.difference(a2_set)
````
