```python
# 모듈 가져오기
import random
# 전역 변수
START_PROMPT = '게임 제목을 입력해 주세요. (최대 28자이내(<=28), 반드시 영문 or 숫자 만 사용 가능)\n'
ERROR_MSG1   = '정확하게 입력하세요'
GAME_TITLE_MAX_LEN = 28
PLAYER_PROMPT = '플레이어의 이름을 입력해 주세요. (최대 28자이내(<=28), 반드시 영문 or 숫자 만 사용 가능)\n'
PLAYER_NAME_MAX_LEN = GAME_TITLE_MAX_LEN
GAME_INTRO_ICON = '*'
GAME_INTRO_SPACE_SIZE = GAME_TITLE_MAX_LEN + 2 # 30
IS_CORRECT = False
RE_TRY = True
AFFIMATIVE_RESPONSE = ['yes', 'y']
NEGATIVE_RESPONSE = ['no', 'n']
TRY_COUNT = 0

#stage 1
def stage1():
  while True:
    # 함수내 지역변수는 함수 안에서만 유효하다!
    # 변수 선언 및 초기화 >> 사용 혹은 업데이트 >> 필요시 반환
    game_title = None
    game_title = input(START_PROMPT).strip()
    if not game_title :
      print(ERROR_MSG1)
    elif len(game_title) > GAME_TITLE_MAX_LEN:
      # 요구사항 추가 -> "왜 않되는지 친철하게 안내라는 메세지 구성"
      print(f'''
        게임 제목을 정확하게 입력하세요.
        최대 {GAME_TITLE_MAX_LEN}자 이내로 입력하시고,
        방금 입력한 내용은 `{game_title}`이고,
        최대 길이 대비 { len(game_title) - GAME_TITLE_MAX_LEN }자가 초과 되었습니다.
      ''')
    else:
      print( '정상입력 완료', game_title)
      break
  return game_title

#stage 2
def stage2():
  player_name = None
  while True:
    player_name = input(PLAYER_PROMPT).strip()
    if not player_name :
      print( ERROR_MSG1 )
      continue
    if len(player_name) > PLAYER_NAME_MAX_LEN:
      print( ERROR_MSG1 )
      continue

    # 정상입력을 수행해야만 진입하는 코드
    break

  return player_name

#stage 3
def stage3(game_title, player_name):
  print( GAME_INTRO_ICON * GAME_INTRO_SPACE_SIZE )
  print( f'{GAME_INTRO_ICON}{game_title:^28}{GAME_INTRO_ICON}')
  print( f'{GAME_INTRO_ICON}{player_name:^28}{GAME_INTRO_ICON}')
  print( f'{GAME_INTRO_ICON}{"press any key":^28}{GAME_INTRO_ICON}')
  print( GAME_INTRO_ICON * GAME_INTRO_SPACE_SIZE )
  pass

#stage 4
def stage4():
  ai_number = random.randint(1,100)
  return ai_number

#stage 5
def stage5():
  while(True):
    player_number = input('숫자 입력 \n')
    print('pppppnum',player_number)
    if( ' ' in player_number ):
      print('공백다시입력')
      continue

    if player_number.startswith('-'):
      if len(player_number) > 1 and player_number[1:].isdigit():
        player_number = int(player_number)
      else:
        print('정수만 입력됩니다.')
        continue
    elif player_number.isdigit():
        player_number = int(player_number)
    else:
      print('정수만 입력됩니다.')
      continue

    if( int(player_number) < 1 or int(player_number) > 100 ):
      print('유효 범위값만 입력됩니다. 1~100')
      continue
    else:
      #print('stage 6 코드에서 보자')
      break

  return player_number


#stage 6
def stage6(player_number, ai_number):
  global IS_CORRECT
  global TRY_COUNT
  while(True):
    #player_number = input()
    TRY_COUNT += 1
    print(f'ai숫: {ai_number} pl숫 :{player_number}')
    if( int(player_number) < 1 or int(player_number) > 100 ):
      print('유효 범위값만 입력됩니다. 1~100')
      break
    elif (player_number > ai_number):
      print('정답은 이거보다 작다')
      break
    elif (player_number < ai_number):
      print('정답은 이거보다 크다')
      break

    if( ai_number == int(player_number) ):
      IS_CORRECT = True
      print(f'정답 // 트라이횟수 : {TRY_COUNT} // 점수 : {(10 - TRY_COUNT)*10}')
      break

#stage 7
def stage7():
  global IS_CORRECT
  while(True):
    game_str=input('다시 할꺼면 y 안할꺼면 n 입력 \n').strip().lower()
    if game_str in AFFIMATIVE_RESPONSE:
      print('스테이지4고고 1// 다시시작!')
      IS_CORRECT = False
      return True
      break
    elif game_str in NEGATIVE_RESPONSE :
      print('ㅂㅂㅂㅂ')
      return False
      break
    else:
      print('다시입력해')
      continue


def main():
  '''
    게임 메인, 전체 스테이지를 구성(조립)
  '''
  '''
  game_title = None
  player_name = None
  game_title = stage1()ㅂ
  player_name = stage2()
  '''
  global RE_TRY
  global IS_CORRECT
  # 함수 호출을 통해서 제목과 이름을 획득
  game_title, player_name = stage1(), stage2()
  # stage3은 이름과 제목을 사용 >> 함수의 입력을 통해서 함수 내부로 전달
  stage3(game_title, player_name)
  #... 하위는 흐름제어에 맞춰서 함수를 정의 배치

  #stage 4~7까지 구성(함수 내부, 현 라인 하위코드)
  #스테이지4
  while(RE_TRY):
    ai_number = stage4()
    while(True):
      player_number = stage5()
      stage6(player_number, ai_number)
      if IS_CORRECT:
        break

    RE_TRY = stage7()
    #print(f'asdsad: {player_number} \n')

if __name__ == '__main__': # 이 코드(혹은 셀, 모듈)의 시작점으로 설정
  main()


```
